```python
path = './dataset/BC5CDR-chem/'
data_list = ['train', 'dev', 'test']
df_all = './dataset/BC5CDR-chem/train.jsonl'
labels = []
for subset in data_list:
    file_path = path+subset+'.jsonl'
    df = pd.read_json(file_path, lines=True)
    labels += list(df.labels.values)
```

```python
ncbi = '../../LLM-IE/few-shot-bioIE/data/BC5CDR-disease/'
data_list = ['train', 'dev', 'test']
for subset in data_list:
    file_path = ncbi + subset + '.csv'
    split_set = pd.read_csv(file_path, sep='\t')

    tokens = list(split_set.sents.values)
    tokens = [sent.split(' ') for sent in tokens]

    labels = list(split_set.ner_seq.values)
    labels = [sent.split(' ') for sent in labels]
    data = {'tokens': tokens, 'labels': labels}
    df = pd.DataFrame(data=data)
    df.to_json(path_or_buf=f'./dataset/BC5CDR-disease/{subset}.jsonl', orient='records', lines=True)
```
